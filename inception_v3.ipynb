```python
from keras.applications import inception_v3
```

```python
# load and download Model 
model=inception_v3.InceptionV3(weights='imagenet')
```

```python
from keras.preprocessing import image
img = image.load_img("lion.jpg", target_size=(299,299))
```

```python
img
```

```python
# change image dimension and type
img_np = image.img_to_array(img)
import numpy as np
ae = np.expand_dims(img_np, axis=0)
```

```python
# let's do some image processing 
from keras.applications.inception_v3 import preprocess_input
finalimg = preprocess_input(ae)
```

```python
# finaly predict image
pred = model.predict(finalimg)
```

```python
# decode prediction
from keras.applications.inception_v3 import decode_predictions
decode_predictions(pred,top=3)
```
